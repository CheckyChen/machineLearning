```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-1,1,100) # 生成有00个-1到1的点
# print(x)
y1 = 2*x + 1

# 自定义一个图例
l1, = plt.plot(x,y1,color='red',linewidth=1.0,linestyle='-')
plt.legend(handles=[l1,],labels=['test1','test2'],loc='best')

ax = plt.gca()

# 将右边的边框设置为空
ax.spines['right'].set_color('none')

# 将上边的边框设置为空
ax.spines['top'].set_color('none')

# 将x轴的刻度设置为 ‘bottom’
ax.xaxis.set_ticks_position('bottom')
# 将y轴的刻度设置为 ‘right’
ax.yaxis.set_ticks_position('left')

# 设置bottom 对应到0点
ax.spines["bottom"].set_position(('data',0))
# 设置right 对应到0点
ax.spines["left"].set_position(('data',0))

x0 = 0.5
y0 = 2*x0 + 1

# 画一个点 (x0,y0),半径 = 50 颜色为蓝色
plt.scatter(x0,y0,s=50,color='b')

# 画一条虚线，从（x0,y0）到（0.5,0），颜色黑色，线宽 2 
plt.plot([x0,x0],[y0,0],'k--',lw=2)

plt.annotate(r'$2x+1=%s'%y0,xy=(x0,y0),xytext=(+30,-30),textcoords='offset points',fontsize=16,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

plt.text(-1,2,r'$this\ is\ the\ text$',fontdict={'size':'16','color':'r'})

plt.show()
```
